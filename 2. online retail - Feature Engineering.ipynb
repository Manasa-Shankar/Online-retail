# Customer segmentation based on:
1. Recency : Number of days since last purchase
2. Frequency : Number of tracsactions
3. Monetary : Total amount of transactions (revenue contributed)

**Monetary**

```python
rfm_m=data.groupby('CustomerID')['Amount'].sum()
rfm_m=rfm_m.reset_index()
rfm_m
```

**Frequency**

```python
rfm_f=data.groupby('CustomerID')['InvoiceNo'].count()
rfm_f=rfm_f.reset_index()
rfm_f
```

**Recency**

```python
rec_date=max(data['InvoiceDate'])
data['Difference']=rec_date-data['InvoiceDate']
data
```

```python
rfm_r=data.groupby('CustomerID')['Difference'].min()
rfm_r=rfm_r.reset_index()
rfm_r
```

```python
rfm_r['Difference']=rfm_r['Difference'].dt.days
```

```python
rfm_r
```

**Combining all the three features into one single dataframe**

```python
rfm=rfm_m.merge(rfm_f,how='inner',on='CustomerID')
rfm=rfm.merge(rfm_r,how='inner',on='CustomerID')
rfm.columns=['CustomerID','Monetary','Frequency','Recency']
rfm
```

# Customer segmentation based on RFM

```python
rfm.describe()
```

```python
rfm.nunique()
```

```python
# 5 is best score and 1 being the worst
rfm['Monetary_score']=pd.qcut(rfm['Monetary'], 5, labels = [1, 2, 3, 4, 5])
rfm['Frequency_score']=pd.qcut(rfm['Frequency'].rank(method="first"),5,labels=[1,2,3,4,5])
rfm['Recency_score']=pd.qcut(rfm['Recency'].rank(method="first"),5,labels=[5,4,3,2,1])
rfm['rfm_score']=rfm['Recency_score'].astype(str)+rfm['Frequency_score'].astype(str)+rfm['Monetary_score'].astype(str)
rfm
```

```python
rfm['rfm_score']=rfm['rfm_score'].astype(int)
rfm['Status']=np.zeros((4338,1))
```

```python
for i in range(0,4338):
    if rfm['rfm_score'][i] in [111, 112, 121, 131,141,151,142]:
        rfm['Status'][i]='Lost customers'
        
    if rfm['rfm_score'][i] in [332, 322, 233, 232, 223, 222, 132, 123, 122, 212, 211]:
        rfm['Status'][i]='Hibernating customers'
        
    if rfm['rfm_score'][i] in [155, 154, 144, 214,215,115, 114, 113]:
        rfm['Status'][i]='Cannot Lose Them'
        
    if rfm['rfm_score'][i] in [255, 254, 245, 244, 253, 252, 243, 242, 235, 234, 225, 224, 153, 152, 145, 143, 135, 134, 133, 125, 124]:
        rfm['Status'][i]='At Risk'
        
    if rfm['rfm_score'][i] in [331, 321, 312, 221, 213, 231, 241, 251]:
        rfm['Status'][i]='About To Sleep'
        
    if rfm['rfm_score'][i] in [535, 534, 443, 434, 343, 334, 325, 324]:
        rfm['Status'][i]='Need Attention'
        
    if rfm['rfm_score'][i] in [525, 524, 523, 522, 521, 515, 514, 513, 425,424, 413,414,415, 315, 314, 313]:
        rfm['Status'][i]='Promising'
        
    if rfm['rfm_score'][i] in [512, 511, 422, 421, 412, 411, 311]:
        rfm['Status'][i]='New Customers'
        
    if rfm['rfm_score'][i] in [553, 551, 552, 541, 542, 533, 532, 531, 452, 451, 442, 441, 431, 453, 433, 432, 423, 353, 352, 351, 342, 341, 333, 323]:
        rfm['Status'][i]='Potential Loyalist'
        
    if rfm['rfm_score'][i] in [543, 444, 435, 355, 354, 345, 344, 335]:
        rfm['Status'][i]='Loyal'
        
    if rfm['rfm_score'][i] in [555, 554, 544, 545, 454, 455, 445]:
        rfm['Status'][i]='Champions'
```

```python
status=pd.DataFrame(rfm.groupby('Status')['CustomerID'].count())
status.reset_index(inplace=True)
status['Percentage']=np.zeros((11,1))
for i in range (0,11):
    status['Percentage'][i]=(status['CustomerID'][i]/np.sum(status['CustomerID']))*100
status.columns=['Status','Number of Customers','Percentage of customers']
```
